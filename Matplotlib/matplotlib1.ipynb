```python
# plotting multiple plots

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels,title

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit VS Virat IPL Run Comparsion")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',marker="*",linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color="#0427D7",linestyle="solid",marker="o")
```
