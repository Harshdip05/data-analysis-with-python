### dtype

```python
int1 = np.array([1,2,3])
print(int1)
print(int1.dtype)

float1 = np.array([1,2,3],dtype=float)
print(float1)
print(float1.dtype)

string1 = np.array(["a","b","c"])
print(string1)
print(string1.dtype)

bool1 = np.array([1,2,3],dtype=bool)
print(bool1)
print(bool1.dtype)

complex1 = np.array([1,2,3],dtype=complex)
print(complex1)
print(complex1.dtype)
```

```python
int8 = np.array([1,2,3],dtype=np.int8)
print(int8)
print(int8.dtype)

int32 = np.array([1,2,3],dtype=np.int32)
print(int32)
print(int32.dtype)
```

### Attributes are properties of Numpy array

```python
a1 = np.arange(10,dtype=np.int32)              ## vector
a2 = np.arange(12,dtype=float).reshape(3,4)    ## matrices
a3 = np.arange(8).reshape(2,2,2)               ## tensor
print(a1)
print(a2)
print(a3)
```

```python
# ndim = no of dimensions

print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
# shape = structure of array

print(a1.shape)   ## 10 elements in one dimension
print(a2.shape)
print(a3.shape)
```

```python
# size = total elements

print(a1.size)
print(a2.size)
print(a3.size)
```

```python
# itemsize = memory per element

print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
# nbytes = total memory

print(a1.nbytes)
print(a2.nbytes)
print(a3.nbytes)
```

```python
# dtype = data type 

print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

#### It is similar to Python’s range() function, but it returns a NumPy array instead of a range object.

```python
arr = np.arange(5)   ##(0,5,1)
print(arr)
```

```python
np.arange(1,11,2)
```

```python
np.arange(0,2,0.5)      ## 0.5 is step size which generate decimal values
```

### reshape

```python
a1 = np.arange(1,11).reshape(2,5)  ## 2rows 5 columns
print(a1)
a2 = np.arange(1,11).reshape(5,2) 
print(a2)
```

```python
b = np.arange(1,13)
print(b,b.size)

b1 = np.arange(1,13).reshape(3,4)
print(b1)
b2 = np.arange(1,13).reshape(4,3)
print(b2)
b3 = np.arange(1,13).reshape(6,2)
print(b3)
b4 = np.arange(1,13).reshape(2,6)
print(b4)
b4 = np.arange(1,13).reshape(5,5)    ## 5*5 = 25 but we have 12 elements only
print(b4)                            ## 5*5 must be equal to array.size 
```

```python
## using -1(automatic size)
# numpy can automatically calculate one dimension

arr = np.array([1,2,3,4,5,6])
print(arr,arr.shape,arr.size)
new_arr = arr.reshape(2,-1)
print(new_arr)
print(new_arr.shape)
print(new_arr.size)
```

```python
arr2 = np.arange(24)
print(arr2,arr2.shape,arr2.size)
new = arr2.reshape(2,3,4)   ## 2matrices 3rows 4 columns
print(new,new.shape,new.size)

arr2 = np.arange(24)
print(arr2,arr2.shape,arr2.size)
new = arr2.reshape(2,3,-1)   ## 2matrices 3rows 4 columns
print(new,new.shape,new.size)
```

```python
## reshape() changes shape not data

## id() tells where the object is stored in memory

arr = np.arange(6)
print(arr,arr.shape,arr.size)
print(id(arr))

new_arr = arr.reshape(2,3)
print(new_arr,new_arr.shape,new_arr.size)
print(id(new_arr))

# id of arr not equal to new_arr
# even though ids are different but both arrays use same data

new_arr[0,0] = 100
print(arr)
print(new_arr)

# original array changed too 
# because both arrays are using the same memory data
```
