```python
## reshape() changes shape not data

## id() tells where the object is stored in memory

arr = np.arange(6)
print(arr,arr.shape,arr.size)
print(id(arr))

new_arr = arr.reshape(2,3)
print(new_arr,new_arr.shape,new_arr.size)
print(id(new_arr))

# id of arr not equal to new_arr
# even though ids are different but both arrays use same data

# id(arr) and id(new_arr) are different because they are different Python objects.
# However, they share the same underlying memory buffer.

new_arr[0,0] = 100
print(arr)
print(new_arr)

print(id(arr))
print(id(new_arr))

# original array changed too 
# because both arrays are using the same memory data
```

```python
# view() creates a new array object but shares the same memory data.
# If you modify one array, the other changes too.

arr = np.array([1,2,3,4])
view_arr = arr.view()

print(arr,id(arr))
print(view_arr,id(view_arr))

view_arr[0] = 100

print(arr,id(arr))
print(view_arr,id(new_arr))

print(view_arr.base is arr)
```

```python
# copy() creates a completely separate array with its own memory.
# Changing one does not affect the other.

arr = np.array([1,2,3,4])
copy_arr = arr.copy()

copy_arr[0] = 100

print(arr,id(arr))
print(copy_arr,id(copy_arr))

print(copy_arr.base is arr)
```

### astype()
astype() creates a new array, so the original array is NOT changed

astype() creates a copy, not a view.

```python
arr = np.array([1,2,3,4])

new_arr = arr.astype(float)

print(arr,id(arr))
print(arr.dtype)

print(new_arr,id(new_arr))
print(new_arr.dtype)

new_arr[0] = 10.5

print(arr)         ## Proof That Original Array Doesn't Change
print(new_arr)
```

```python
arr_int64 = np.array([1,2,3])
print(arr_int64,arr_int64.dtype)         ## 8 bytes per element

new_int16 = arr_int64.astype(np.int16)
print(new_int16,new_int16.dtype)         ## 2 bytes per element

# reduce memory usage
# improve efficiency
# match required datatype
```

| Feature     | `ndenumerate` 🟢 | `nditer` 🔵       |
| ----------- | ---------------- | ----------------- |
| Output      | (index, value)   | value only        |
| Speed       | Slower           | Faster ⚡          |
| Control     | Simple           | Advanced          |
| Use case    | Debugging        | Performance tasks |
| Multi-array | ❌ No             | ✅ Yes             |


```python
# ndenumerate

import numpy as np

arr = np.array([[10, 20],
                [30, 40]])

for index, value in np.ndenumerate(arr):
    print(index, value)
```

```python
arr = np.array([[[1,2],[3,4]]])

for idx, val in np.ndenumerate(arr):
    print(idx, val)
```

| ------------ | ---------------------------- | ------------------------ |
