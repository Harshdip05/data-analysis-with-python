### Arithmetic operators

```python
# NumPy arithmetic =
# “Apply operation to every element automatically”
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a+b)
print(a-b)
print(a/b)
print(a*b)
print(a%b)
print(a//b)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print(a+b)
print(a-b)
print(a/b)
print(a*b)
print(a%b)
print(a//b)
```

### Comparison operators

```python
# Comparison operators are used to compare values in arrays and return True / False.
# 👉 In NumPy, they work element-wise (on each element separately)


# Comparison operators in NumPy:
# Work element-wise
# Return boolean arrays
# Used for filtering and conditions
```

```python
import numpy as np

a = np.array([1, 2, 3, 4])
print(a)

a>1
a==1
a[a > 2]  ## boolean indexing
a[(a > 1) & (a < 4)]  ## Combining Conditions
```

### Logical operators

```python
# In NumPy:
        # NEVER use and, or, not
        # ALWAYS use &, |, ~


# Logical operators are used to combine conditions (True/False arrays).
# 👉 They are mostly used in boolean indexing (filtering data)
```

```python
import numpy as np

a = np.array([1, 2, 3, 4])

a[(a > 1) & (a < 4)]

a[(a < 2) | (a > 3)]

a[~(a > 2)]


# Always Use Brackets
# # ✅ Correct
# (a > 1) & (a < 4)
# # ❌ Wrong
# a > 1 & a < 4
```

### Assignment operators

```python
# Assignment operators are used to update values in an array.
# 👉 They work element-wise (just like other NumPy operations)
```

```python
import numpy as np

a = np.array([1, 2, 3])
a = np.array([10, 20, 30])   ## Basic Assignment (=)
print(a)
```

```python
a = np.array([1, 2, 3])
a += 5
print(a)
```

```python
a = np.array([1, 2, 3])
a *= 3
print(a)

a = np.array([5, 6, 7])
a //= 2
print(a)
```

### Special NumPy Operations (Functions)

```python
# These are built-in NumPy functions that make calculations faster and easier than normal Python.
# 👉 Think of them as ready-made powerful tools
```

```python
import numpy as np

a = np.array([1, 4, 9])
print(a)

np.sqrt(a)
np.exp(a)
np.log(a)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

np.add(a, b)        # same as a + b
np.subtract(a, b)   # same as a - b
np.multiply(a, b)   # same as a * b
np.divide(a, b)     # same as a / b
```

```python
a = np.array([1, 2, 3, 4])

np.mean(a)
np.std(a)
np.max(a)
np.min(a)
```

### Matrix Operators

```python
# Matrix operations are used when working with 2D arrays (rows × columns).
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

A + B
A - B
A * B  ## Element-wise Multiplication, Not real matrix multiplication!
```

```python
A @ B   ## Matrix Multiplication

# For matrix multiplication: (columns of A) = (rows of B)
```

```python
print(np.dot(A,B))   ## Same as A @ B
```
